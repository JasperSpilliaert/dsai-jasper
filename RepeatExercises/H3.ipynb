### 1. z-tests:

Z-tests worden gebruikt wanneer we de populatievariantie kennen en de steekproefomvang groot is (meestal n > 30). Ze worden gebruikt om hypothesen te testen over de populatiegemiddelde.

-	Left-tailed: We willen testen of de gemiddelde lengte van een groep studenten kleiner is dan 170 cm.
    - H0: Œº ‚â• 170 cm (nulhypothese)
    - H1: Œº < 170 cm (alternatieve hypothese)
    - Wanneer te gebruiken: Gebruik dit wanneer je vermoedt dat de werkelijke populatiegemiddelde kleiner is dan een bepaalde waarde.
-	Right-tailed: We willen testen of de gemiddelde opbrengst van een bepaald gewas groter is dan 500 kg per hectare.
    - H0: Œº ‚â§ 500 kg/ha
    - H1: Œº > 500 kg/ha
    - Wanneer te gebruiken: Gebruik dit wanneer je vermoedt dat de werkelijke populatiegemiddelde groter is dan een bepaalde waarde.
-	Two-tailed: We willen testen of de gemiddelde IQ-score van een populatie verschilt van 100.
    - H0: Œº = 100Œº
    - H1: Œº ‚â† 100Œº
    - Wanneer te gebruiken: Gebruik dit wanneer je vermoedt dat de werkelijke populatiegemiddelde anders is dan een bepaalde waarde, zonder vooraf te bepalen of het hoger of lager is.
### 2. t-test:

T-tests worden gebruikt wanneer de populatievariantie niet bekend is en de steekproefomvang klein is (meestal n‚â§30n‚â§30). Ze worden gebruikt om hypothesen te testen over de populatiegemiddelde.

-	Left-tailed: Voorbeeld: We willen testen of de gemiddelde score op een examen kleiner is dan 50 bij een kleine steekproef van studenten.
    - H0: Œº ‚â• 50
    - H1: Œº < 50
    - Wanneer te gebruiken: Gebruik dit wanneer je vermoedt dat de werkelijke populatiegemiddelde kleiner is dan een bepaalde waarde, en je hebt een kleine steekproef.

-	Right-tailed: Voorbeeld: We willen testen of de gemiddelde tijd die een kleine groep pati√´nten besteedt aan lichaamsbeweging meer dan 30 minuten per dag is.
    - H0: Œº ‚â§ 30
    - H1: Œº > 30
    - Wanneer te gebruiken: Gebruik dit wanneer je vermoedt dat de werkelijke populatiegemiddelde groter is dan een bepaalde waarde, en je hebt een kleine steekproef.

-	Two-tailed: Voorbeeld: We willen testen of de gemiddelde bloeddruk van een kleine steekproef van pati√´nten verschilt van 120 mmHg.
    - H0: Œº = 120
    - H1: Œº ‚â† 120
    - Wanneer te gebruiken: Gebruik dit wanneer je vermoedt dat de werkelijke populatiegemiddelde anders is dan een bepaalde waarde, zonder vooraf te bepalen of het hoger of lager is, en je hebt een kleine steekproef.

- Nulhypothese (H‚ÇÄ): Dit is de hypothese die je wilt testen. Meestal stelt de nulhypothese dat er geen effect of geen verschil is.
- Alternatieve hypothese (H‚ÇÅ of Ha): Dit is de hypothese die je overweegt als de nulhypothese onwaar blijkt te zijn.

- Wat gebeurt er tijdens een hypothesetest?

    - Verwerpen van de nulhypothese (H‚ÇÄ): Als de testresultaten significant zijn (bijvoorbeeld als de p-waarde kleiner is dan het gekozen significatieniveau, vaak 0,05), dan verwerp je de nulhypothese. Dit betekent dat er voldoende bewijs is om te concluderen dat de alternatieve hypothese waarschijnlijker is.
    - Niet verwerpen van de nulhypothese (H‚ÇÄ): Als de testresultaten niet significant zijn, dan verwerp je de nulhypothese niet. Dit betekent dat er niet genoeg bewijs is om te zeggen dat de alternatieve hypothese waar is. Dit is echter geen bewijs dat de nulhypothese waar is; het betekent alleen dat je op basis van de gegevens de nulhypothese niet kunt verwerpen.

When drawing a random sample consisting of 50 observations from a population with variance ùúé2 = 55 we obtain as sample mean ùë• = 25.

We now want to find out if there is a reason to assume that the population mean is smaller than 27.

```python
import numpy as np
from scipy.stats import norm

# Given data
sample_mean = 25
population_variance = 55
sample_size = 50
hypothesized_mean = 27

# Calculate population standard deviation
population_std_dev = np.sqrt(population_variance)

# Calculate the z-statistic
z_statistic = (sample_mean - hypothesized_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate the p-value for a one-tailed test
p_value = norm.cdf(z_statistic)

# Print results
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Determine if the result is significant at the 0.05 level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is significant evidence to suggest that the population mean is smaller than 27.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest that the population mean is smaller than 27.")
```

## Question 4

The diameter of pistons produced in a factory are _D_ cm where _D_ ~ Nor(13.4, 0.12)

(1) The tolerance specified for the pistons is that the diameter needs to be at least 13.35 cm and not more than 13.5 cm. What proportion of the production in the factory meets these tolerance limits?  
(2) Three pistons are chosen at random. What is the probability that none of them meet the tolerance limits  
(3) A sample of 20 pistons is taken at random. The mean value for the diameter is found to be 13.43 cm. Test at the 5% significance level whether any modifications have been made to the machine producing the pistons.


```python
import math

#N(13.4 , 4) = N(mu, sigma^2)
#Nor(13.4 , 3) = Nor(mu , sigma) 

#1
mu = 13.4 
sigma = 0.12
p = stats.norm.cdf(13.5, loc = mu,scale = sigma) - stats.norm.cdf(13.35, loc = mu,scale = sigma)
print(p)

#2
print((1-p)**3)
# 2.1 wat is de kans dat tenminste 1 voldoet: 
# print(1 - (1-p)**3)

#3 Hypothesis testing
sample_mean = 13.43
n = 20

SE = sigma / (n ** 0.5)
Z = (sample_mean - mu) / SE
print(Z)

# Testing the Mean Diameter of 20 Pistons: The Z-value remains approximately 0.387. 
# As this value is less than the critical Z-value of 1.96 for a two-tailed test at the 5% significance level,  we still do not reject the null hypothesis, 
# indicating insufficient evidence to suggest a change in the mean diameter of the pistons from 13.4 cm based on this sample.
```

The testing procedure then runs as follows:

**Step 1.** Formulate the hypotheses:

- $H_0: \mu = 13.4$
- $H_1: \mu > 13.4$

**Step 2.** Choose a significance level, e.g. $\alpha = 0.025$

```python
# Properties of the sample:
n = 20      # sample size
sm = 13.43  # sample mean
s = 0.12 # population standard deviation (assumed to be known)
a = 0.05    # significance level (chosen by the researcher)
m0 = 13.4   # hypothetical population mean (H0)

#-------
# PLOT:
#------

# Gauss-curve plot:
# X-values
dist_x = np.linspace(m0 - 4 * s/np.sqrt(n), m0 + 4 * s/np.sqrt(n), num=201)
# Y-values for the Gauss curve
dist_y = stats.norm.pdf(dist_x, m0, s/np.sqrt(n))
fig, dplot = plt.subplots(1, 1)
# Plot the Gauss-curve
dplot.plot(dist_x, dist_y)
# Show the hypothetical population mean with an orange line
dplot.axvline(m0, color="orange", lw=2)
# Show the sample mean with a red line
dplot.axvline(sm, color="red")
```

**Step 3.** Determine the value of the test statistic, in this case $\overline{x} = 13.43$

**Step 4.** 

Determine the $p$-value and reject $H_0$ if $p < \alpha$.  
The $p$-value is the probability, if the null hypothesis is true, to obtain
a value for the test statistic that is at least as extreme as the
observed value


```python
p = stats.norm.sf(sm, loc=m0, scale=s/np.sqrt(n))
print("p-value: %.5f" % p)
if(p < a):
    print("p < a: reject H0")
else:
    print("p > a: do not reject H0")
```

## Vraag 6 [5 pt] 

Het gewicht van jongens op een bepaalde leeftijd is normaal verdeeld met een gemiddelde van 37.8 kg. Tegen een weeshuis in een ontwikkelingsland wordt een klacht ingediend door een NGO dat de kinderen die er verblijven ondervoed zijn. Een aselecte steekproef levert de hieronder gegeven metingen op.

Is er reden om te vermoeden dat de jongens in dit weeshuis ondervoed zijn? Gebruik een significantieniveau van 5%.

1. [1pt] Welke hypothesetoets ga je toepassen om deze onderzoeksvraag te beantwoorden? Wees zo specifiek mogelijk!

2. [1pt] Formuleer de nulhypothese en de alternatieve hypothese

3. [1pt] Bereken de gepaste toetsingsgrootheid (teststatistiek) voor deze toets. Geef het symbool en de waarde.

4. [1pt] Bereken de p-waarde.

5. [1pt] Trek een besluit op basis van de vorige stap en formuleer een antwoord op de onderzoeksvraag.

```python
weights = np.array(
    [36.8, 37.3, 37.4, 37.2, 32. , 35.3, 40.5, 35.3, 31.2, 39.6,
     36.1, 29.5, 35. , 37. , 35.6, 35. , 36.8, 38.1, 39.6, 35.6,
     42.3, 38.6, 37.8, 39.2, 39.4])
## LEFT TAIL t-TEST
#Step 1: formulate the null and alternative hypotheses
#- H0: Œº = 100
#- H1: Œº < 100

#Step 2: specify the significance level
# Properties of the sample:
n = len(weights)  # sample size
sm = weights.mean() # sample mean
s = weights.std(ddof=1) # sample standard deviation (assumed to be known)
a = 0.05 # significance level (chosen by the researcher)
m0 = 37.8 # hypothetical population mean (H0)
# Plotting the sample distribution
# Gauss-curve plot:
# X-values
dist_x = np.linspace(m0 - 4 * s/np.sqrt(n), m0 + 4 * s/np.sqrt(n), num=201)
# Y-values for the Gauss curve
dist_y = stats.t.pdf(dist_x, loc= m0, scale= s/np.sqrt(n), df = n-1)
fig, dplot = plt.subplots(1, 1)
# Plot the Gauss-curve
dplot.plot(dist_x, dist_y)
# Show the hypothetical population mean with an orange line
dplot.axvline(m0, color="orange", lw=2)
# Show the sample mean with a red line
dplot.axvline(sm, color="red")

#Step 3: compute the test statistic (red line in the plot)
# Hier is dat 19.94

#Step 4:
## method 1
# Determine the $p$-value and reject $H_0$ if $p < \alpha$.
#The $p$-value is the probability, if the null hypothesis is true, to obtain
#a value for the test statistic that is at least as extreme as the
# observed value
p = stats.t.cdf(sm, loc=m0, scale=s/np.sqrt(n), df=n-1)
print("p-value: %.5f" % p)
if(p < a):
    print("p < a: reject H0")
else:
    print("p > a: do not reject H0")
## method 2
# An alternative method is to determine the critical region, i.e. the set of all values for the sample mean where $H_0$ may be rejected.
# The boundary of that area is called the critical value $g$. To the right of it you can't reject $H_0$ (acceptance region), to the left you can (critical region).
# The area of the acceptance region is $\alpha$, the area of the critical region is $1 - \alpha$.
g = stats.t.isf(1-a, loc = m0, scale = s / np.sqrt(n), df=n-1)
print("Critical value g ‚âÉ %.3f" % g)
if (sm > g):
    print("sample mean = %.3f > g = %.3f: do not reject H0" % (sm, g))
else:
    print("sample mean = %.3f < g = %.3f: reject H0" % (sm, g))

# We can conclude that there is not enough evidence to reject the
# null hypothesis.
```
