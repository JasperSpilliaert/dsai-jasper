## Vraag 3 [5 pt]

Stel dat 60% van alle klanten van een grote verzekeringsmaatschappij een autopolis heeft bij de verzekeringsmaatschappij, 40% heeft een brandpolis voor zijn huis en 25% heeft zowel een autopolis als een brandpolis.
A = de geselecteerde klant heeft een autopolis
B = de geselecteerde klant heeft een brandpolis

1. Zijn A en B onafhankelijke events? Leg kort uit waarom wel of waarom niet?
1. Wat is de kans dat een klant ten minste een autopolis of een brandpolis heeft?
2. Wat is de kans dat een klant geen autopolis en geen brandpolis heeft ?
3. Wat is de kans dat een klant óf een autopolis óf een brandpolis heeft?
4. Wat is de kans dat een klant enkel een brandpolis heeft?

### Antwoord vraag 3

1. Twee gebeurtenissen zijn onafhankelijk als P(E ∩ F) = P(E) * P(F)

P(A ∩ B) = P(A) * P(B)\
         = 0.25 = 0.6 * 0.4\
         = 0.25 != 0.24

gebeurtenissen A en B zijn dus afhankelijk


2. P(A ∪ B) = P(A) + P(B) - P(A ∩ B)\
         = 0.6 + 0.4 - 0.25\
         = 0.75

de kans dat de klant ten minste een autopolis of brandpolis heeft is 75%

3. P(geen polis) = 1 - P(A ∪ B)\
= 1 - 0.75/
= 0.25

de kans dat de klant geen autopolis of brandpolis heeft is 25%



4.
P(A ∩ B') = 0.25

P(A ∪ B') = 0.75

P(A ∪ B') - P(A ∩ B') = 0.5

de kans dat de klant of een brandpolis heeft of een autopolis is 50%


5. P(B zonder A) = P(B) - P(A ∩ B) = 0.4 - 0.25 = 0.15

de kans dat de klant enkel een brandpolis heeft is 15%

```python
m = 17  # Sample mean
s = 4.8    # Sample (!) standard deviation
n = 40       # Sample size
alpha = .1                  # 1 - alpha is the confidence level

# Step 2.
z = stats.norm.isf(alpha/2)
print("z-score: %.5f" % z)

# Step 3.
lo = m - z * s / np.sqrt(n)
hi = m + z * s / np.sqrt(n)
print("Confidence interval: [%.4f, %.4f]" % (lo, hi))
```

```python
kcal = [255, 244, 239, 242, 265, 245, 259, 248, 225, 226, 251, 233,255, 244, 239, 242, 265, 245, 259, 248, 225, 226, 251, 233,255, 244, 239, 242, 265, 245, 259, 248, 225, 226, 251, 233,255, 244, 239, 242, 265, 245, 259, 248, 225, 226, 251, 233]

np.mean(kcal)
```

```python
# Properties of the sample:
n = 40      # sample size
sm = np.mean(kcal)  # sample mean
s = np.std(kcal)    # population standard deviation (assumed to be known)
a = 0.05    # significance level (chosen by the researcher)
mu = 240    # hypothetical population mean (H0)
```

```python
p = stats.t.sf(sm, loc=mu, scale=s/np.sqrt(n) , df= n-1)
print("p-value: %.5f" % p)
if(p < a):
    print("p < a: reject H0")
else:
    print("p > a: do not reject H0")
```

```python
g = stats.t.isf(a, loc=mu, scale=s / np.sqrt(n), df=n-1) #
print("Critical value g ≃ %.3f" % g)
if (sm < g):
    print("sample mean = %.3f < g = %.3f: do not reject H0" % (sm, g))
else:
    print("sample mean = %.3f > g = %.3f: reject H0" % (sm, g))
```

```python
# Gauss-curve plot:
# X-values
dist_x = np.linspace(mu - 4 * s/np.sqrt(n), mu + 4 * s/np.sqrt(n), num=201)
# Y-values
dist_y = stats.norm.pdf(dist_x, loc=mu, scale=s/np.sqrt(n))
fig, dplot = plt.subplots(1, 1)
# Plot
dplot.plot(dist_x, dist_y)
# Hypothetical population mean in orange
dplot.axvline(mu, color="orange", lw=2)
# Sample mean in red
dplot.axvline(sm, color="red")
# Fill the acceptance region in light blue
dplot.fill_between(dist_x, 0, dist_y, where=dist_x <= g, color='lightblue');
```

## conclusie stap 5

er is voldoende aanwijzing om H0 te verwerpen

```python
def detect_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = [value for value in data if value < lower_bound or value > upper_bound]
    return outliers

niet_rokende_outliers = detect_outliers(niet_rokende_moeders)
print("Aantal outliers bij niet-rokende moeders:", len(niet_rokende_outliers))
```
