## 1. Simple Exponential Smoothing (SEM)

- Voorbeeld: Tijdreeksdata zonder trend of seizoensgebondenheid.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing

# Simuleer tijdreeksdata zonder trend of seizoen
np.random.seed(0)
data = np.random.normal(loc=10, scale=2, size=100)  # Data rondom een gemiddelde van 10

# Zet data om in een DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Pas Simple Exponential Smoothing toe
model = ExponentialSmoothing(df['Value'], trend=None, seasonal=None)
fit = model.fit(smoothing_level=0.2)

# Plot de gegevens en de voorspellingen
plt.figure(figsize=(10, 6))
plt.plot(df['Value'], label='Observed Data')
plt.plot(fit.fittedvalues, label='Fitted Values', linestyle='--')
plt.title('Simple Exponential Smoothing')
plt.legend()
plt.show()
```

## 2. Double Exponential Smoothing (DEM)

Voorbeeld: Tijdreeksdata met een trend.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing

# Simuleer tijdreeksdata met trend
np.random.seed(0)
time = np.arange(100)
data = 5 + 0.5 * time + np.random.normal(scale=2, size=100)  # Lineaire trend met ruis

# Zet data om in een DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Pas Double Exponential Smoothing toe
model = ExponentialSmoothing(df['Value'], trend='add', seasonal=None)
fit = model.fit(smoothing_level=0.6, smoothing_slope=0.2)

# Plot de gegevens en de voorspellingen
plt.figure(figsize=(10, 6))
plt.plot(df['Value'], label='Observed Data')
plt.plot(fit.fittedvalues, label='Fitted Values', linestyle='--')
plt.title('Double Exponential Smoothing')
plt.legend()
plt.show()
```

## 3. Triple Exponential Smoothing (TES)

Voorbeeld: Tijdreeksdata met zowel trend als seizoensgebondenheid.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing

# Simuleer tijdreeksdata met trend en seizoensgebondenheid
np.random.seed(0)
time = np.arange(120)
seasonal_effect = 10 + 5 * np.sin(2 * np.pi * time / 12)  # Seizoensgebonden effect
trend_effect = 0.2 * time  # Lineaire trend
data = 15 + trend_effect + seasonal_effect + np.random.normal(scale=2, size=120)  # Gecombineerd effect

# Zet data om in een DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Pas Triple Exponential Smoothing toe
model = ExponentialSmoothing(df['Value'], trend='add', seasonal='add', seasonal_periods=12)
fit = model.fit(smoothing_level=0.6, smoothing_slope=0.2, smoothing_seasonal=0.4)

# Plot de gegevens en de voorspellingen
plt.figure(figsize=(10, 6))
plt.plot(df['Value'], label='Observed Data')
plt.plot(fit.fittedvalues, label='Fitted Values', linestyle='--')
plt.title('Triple Exponential Smoothing')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

train =dfwalmart.omzet

wounded_hw = ExponentialSmoothing(train,
  trend='add', seasonal='add',
  seasonal_periods=4).fit()

wounded_hw.fittedvalues.plot(legend=True, label='fitted')
wounded_predicted = wounded_hw.forecast(12)

train.plot(legend=True, label='train')
wounded_hw.fittedvalues.plot(legend=True, label='fitted')
wounded_predicted.plot(legend=True, label='predicted')

plt.title('Train, test, fitted & predicted values using Holt-Winters')
print(wounded_predicted)

if wounded_predicted[wounded_predicted>100].any():
  print("iets")
```

```python
reached=False
counter=1
while(not reached):
  wounded_predicted = wounded_hw.forecast(counter)
  if wounded_predicted[wounded_predicted>150].any():
    reached=True
    print(wounded_predicted[wounded_predicted>150])
  counter+=1
```
