### 8. Regression !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! -> enige belangrijke van dit hoofdstuk

- Doel: Regressieanalyse onderzoekt de relatie tussen twee of meer kwantitatieve variabelen, waarbij je de invloed van een onafhankelijke variabele op een afhankelijke variabele wilt voorspellen.

    - Voorbeeldvraag: Hoe beïnvloedt het aantal uren studie de eindexamencijfers van studenten?
        - Onafhankelijke variabele (X): Aantal uren studie (kwantitatief)
        - Afhankelijke variabele (Y): Eindexamencijfer (kwantitatief)

    - Je voert een regressieanalyse uit om te bepalen hoe goed het aantal uren studie (onafhankelijke variabele) de eindexamencijfers (afhankelijke variabele) voorspelt. 

### 9.  Covariance

- Doel: Covariantie meet de mate en richting van de gezamenlijke variatie van twee kwantitatieve variabelen.

    - Voorbeeldvraag: Wat is de covariantie tussen het aantal uren studie en eindexamencijfers?
        - Analyse: Als de covariantie positief is, betekent dit dat als het aantal uren studie toeneemt, de eindexamencijfers ook waarschijnlijk toenemen. Een negatieve covariantie zou het tegenovergestelde betekenen.

### 10. Pearson’s Correlation Coefficient (r)

- Doel: Pearson's correlatiecoëfficiënt meet de sterkte en richting van de lineaire relatie tussen twee kwantitatieve variabelen. De waarde ligt tussen -1 en +1.

    - Voorbeeldvraag: Wat is de Pearson correlatiecoëfficiënt tussen het aantal uren studie en eindexamencijfers?
        - Analyse: Een Pearson’s rr waarde van +0.85 zou een sterke positieve lineaire relatie aangeven, terwijl een waarde van -0.60 een sterke negatieve lineaire relatie zou aanduiden.

### 11. Coefficient of Determination (R²)

- Doel: Het determinatiecoëfficiënt geeft aan welk percentage van de variatie in de afhankelijke variabele kan worden verklaard door de onafhankelijke variabele(s) in een regressiemodel. Het is de kwadraat van Pearson’s correlatiecoëfficiënt voor een eenvoudige regressie.

    - Voorbeeldvraag: Wat is het R2R2 van de regressieanalyse tussen het aantal uren studie en eindexamencijfers?
        - Analyse: Als R2R2 = 0.72, betekent dit dat 72% van de variabiliteit in de eindexamencijfers verklaard wordt door het aantal uren studie.

### 12. Correlation

- Doel: Correlatieanalyse onderzoekt de sterkte en richting van de relatie tussen twee kwantitatieve variabelen.

    - Voorbeeldvraag: Wat is de correlatie tussen het aantal uren studie en eindexamencijfers?
        - Analyse: Hier bereken je de Pearson’s correlatiecoëfficiënt om te bepalen hoe sterk en in welke richting het aantal uren studie samenhangt met de eindexamencijfers.

2 kwantitatieve variabelen

- Regression: Onderzoekt hoe de verandering in een onafhankelijke kwantitatieve variabele een afhankelijke kwantitatieve variabele beïnvloedt.

- Voorbeeldvraag: Hoe beïnvloedt het aantal uren studie de eindexamencijfers van studenten?
    - Onafhankelijke variabele (X): Aantal uren studie (kwantitatief)
    - Afhankelijke variabele (Y): Eindexamencijfer (kwantitatief)


- Covariance: Meet de gezamenlijke variatie van twee kwantitatieve variabelen, maar geeft geen gestandaardiseerde maat.
- Pearson’s Correlation Coefficient: Meet de sterkte en richting van de lineaire relatie tussen twee kwantitatieve variabelen.
- Coefficient of Determination (R²): Geeft aan welk percentage van de variatie in de afhankelijke variabele verklaard wordt door de onafhankelijke variabele(s).
- Correlation: Algemeen begrip dat de sterkte en richting van de relatie tussen twee kwantitatieve variabelen onderzoekt, vaak gemeten door Pearson's correlatiecoëfficiënt.
